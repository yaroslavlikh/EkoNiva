### Исследовательский вопрос 2: "Влияние различных технологий выращивания кормовых культур на их питательную ценность"

#### Цели:
- Исследовать, как различные технологии выращивания (из fields.csv) влияют на питательную ценность кормовых культур.

#### Методы:

1. Анализ питательных веществ в почве:
   - Сравнить уровни p2o5, k2o, гумуса и pH для полей с разными технологиями выращивания кормовых культур.

2. Оценка питательной ценности урожая:
   - Используя данные из works.csv (содержание сухого вещества и влажности), рассчитать питательную ценность урожая для каждой культуры и технологии.

3. Корреляция с молочной продукцией (косвенно):
   - Если есть возможность, сопоставить данные о питательной ценности кормов с косвенными показателями молочной продуктивности (например, динамика затрат на корма в молочном производстве).

Эти исследования позволят:

- Понять значимость затрат на удобрения для молочного производства ЭкоНивы через влияние на урожайность и качество кормовых культур.
- Определить оптимальные технологии выращивания для повышения эффективности молочного производства.
- Выявить возможности для снижения затрат и повышения качества кормов без ущерба для молочной продуктивности.
